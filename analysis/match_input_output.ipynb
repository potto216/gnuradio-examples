```python
import numpy as np
import matplotlib.pyplot as plt
import wave
import sys

# Open the wav file
file_path = '~/data/Exercise_3d_PAM_Rx_RRC.wav'
wav_file = wave.open(file_path, 'r')

# Extract Raw Audio from Wav File
signal = wav_file.readframes(-1)
signal = np.frombuffer(signal, dtype=np.int16)

# Get the number of channels
channels = wav_file.getnchannels()

# Split the data into channels
if channels == 2:
    signal = np.reshape(signal, (-1, 2))
    channel_1 = signal[:, 0]
    channel_2 = signal[:, 1]
else:
    print('The file does not have 2 channels')
    sys.exit(0)

# Plot the two channels against each other
plt.figure(figsize=(10, 6))
plt.plot(channel_1, label='Channel 1')
plt.plot(channel_2, label='Channel 2')
plt.legend()
plt.title('Channels Plot')
plt.xlabel('Sample Index')
plt.ylabel('Amplitude')
plt.show()

# Find the delay using correlation
correlation = np.correlate(channel_1, channel_2, mode='full')
delay = np.argmax(correlation) - (len(channel_1) - 1)
print(f'Delay between channels: {delay} samples')
```

```python
import numpy as np
import plotly.graph_objects as go
# import soundfile as sf
import wave
import sys

# Open the wav file
file_path = '/home/user/data/Exercise_3d_PAM_Rx_RRC.wav'
                    

# Open the wav file
wav_file = wave.open(file_path, 'r')

# Extract Raw Audio from Wav File
signal = wav_file.readframes(-1)
signal = np.frombuffer(signal, dtype=np.int16)

# Get the number of channels
channels = wav_file.getnchannels()

# Split the data into channels
if channels == 2:
    signal = np.reshape(signal, (-1, 2))
    channel_1 = signal[:, 0]
    channel_2 = signal[:, 1]
else:
    print('The file does not have 2 channels')
    sys.exit(0)


# Create interactive plot
fig = go.Figure()
fig.add_trace(go.Scatter(y=channel_1, mode='lines', name='Channel 1'))
fig.add_trace(go.Scatter(y=channel_2, mode='lines', name='Channel 2'))
fig.update_layout(title='Channels Plot', xaxis_title='Sample Index', yaxis_title='Amplitude')
fig.show()

# Find the delay using correlation
correlation = np.correlate(channel_1, channel_2, mode='full')
delay = np.argmax(correlation) - (len(channel_1) - 1)
print(f'Delay between channels: {delay} samples')

# Create interactive plot
fig = go.Figure()
fig.add_trace(go.Scatter(y=channel_1[delay:], mode='lines', name='Channel 1'))
fig.add_trace(go.Scatter(y=channel_2, mode='lines', name='Channel 2'))
fig.update_layout(title='Aligned Channels Plot', xaxis_title='Sample Index', yaxis_title='Amplitude')
fig.show()
```
